### Function

```python
# Define a function

def int2hex (n):
  return ('0x%0.2X' % n)[2:]

int2hex(255)
```

```python
# Define a function with named arguments

def rgb2hex (red=0, green=0, blue=0):
  return '#' + int2hex(red) + int2hex(green) + int2hex(blue)

print(rgb2hex(211))
print(rgb2hex(green=255))
```

### List/dictionary comprehension

```python
# Make a list by looping through another list
# *color is spread operator, equivalent to rgb2hex(color[0], color[1], color[2])

colors = [(255, 0, 0), (0, 255, 0), (0, 0, 255)]

[rgb2hex(*color) for color in colors]
```

```python
# Make a dictionary by looping through another list

color_names = ['red', 'green', 'blue']

color_dictionary = {color_names[i]:rgb2hex(*colors[i]) for i in range(len(colors))}

color_dictionary['red']
```

### Dimension reduction with scikit-learn

```python
from sklearn.manifold import TSNE, MDS
import altair as alt
import scipy
from sklearn.decomposition import PCA
import numpy as np
import seaborn as sns
```

```python
# Extract values from dataframe

X = df_pokemon[['attack', 'defense', 'sp_attack', 'sp_defense', 'speed', 'hp']].values
```

```python
mean_X = np.mean(X,axis = 0)
std = np.std(X,axis = 0)
```

```python
X_normed = (X - mean_X)/std
X_normed
s,v,ut = scipy.linalg.svd(X_normed)
```

```python
eigvectors = ut[:2,:]
reduced_dimension_vector = np.matmul(eigvectors,np.transpose(X_normed))
```

```python
pca_df = pd.DataFrame(data = np.transpose(reduced_dimension_vector),columns=["PCA1","PCA2"])
pca_df["is_legendary"] = df_pokemon["is_legendary"]
```

```python
fig,ax = plt.subplots(1,1,figsize = (12,12))
sns.scatterplot(data = pca_df,x = "PCA1",y = "PCA2",hue = "is_legendary",ax = ax)
```

```python
# t-sne algorithm, 
# t-sne is a randomized algorithm, the argument random_state is the random seed,
# it is set to 801 for reproducibile plotting, you may change the seed and see 
# how the plot changes

X_embedded_30 = TSNE(n_components=2, random_state=801,perplexity = 30).fit_transform(X)
X_embedded_50 = TSNE(n_components=2, random_state=801,perplexity = 50).fit_transform(X)
X_embedded_5 = TSNE(n_components=2, random_state=801,perplexity = 5).fit_transform(X)
```

```python
# Concat results to dataframe

df_pokemon['x_30'] = X_embedded_30[:, 0]
df_pokemon['y_30'] = X_embedded_30[:, 1]
df_pokemon['x_5'] = X_embedded_5[:, 0]
df_pokemon['y_5'] = X_embedded_5[:, 1]
df_pokemon['x_50'] = X_embedded_50[:, 0]
df_pokemon['y_50'] = X_embedded_50[:, 1]
```

```python
# Plot with altair

alt.Chart(df_pokemon, title='Pokemon base stats t-sne').mark_point().encode(
    x='x_5',
    y='y_5',
    color='is_legendary:N',
    tooltip=['name', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed', 'hp', 'is_legendary', 'base_total']
)
```
